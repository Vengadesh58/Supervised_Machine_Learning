```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
def get_data_fun(url):

    """ This function will fecth the data from api url and return the data"""

    import requests
    from requests.packages.urllib3.exceptions import InsecureRequestWarning

    # Disable SSL certificate verification warning
    requests.packages.urllib3.disable_warnings(InsecureRequestWarning)

    # Make a GET request to the API endpoint
    response = requests.get(url,verify=False)

    # Check the response status code
    if response.status_code == 200:
        # Request was successful
        data = response.json()
        # Process the data as needed
        print(data)
        return data
    else:
        # Request failed
        print('Error:', response.status_code)
```

```python
top_team = get_data_fun("https://vlgspfitcdx.devsys.net.sap:4430/dashboard/topTeam")
```

```python
def dataframe(data):

    """ this function will convert the json data into dataframe"""
    df = pd.DataFrame(data)
    print(df.head())
    return df
```

```python
top_team_df = dataframe(top_team)
top_team_df.to_csv('top_team.cvs',index=False)
```

```python
def get_data_fun_params(url,param):

    """ This function will fecth the data from api url and return the data"""

    import requests
    from requests.packages.urllib3.exceptions import InsecureRequestWarning

    # Disable SSL certificate verification warning
    requests.packages.urllib3.disable_warnings(InsecureRequestWarning)

    params ={
        'activity':param
    }

    # Make a GET request to the API endpoint
    response = requests.get(url,verify=False,params=params)

    # Check the response status code
    if response.status_code == 200:
        # Request was successful
        data = response.json()
        # Process the data as needed
        return data
    else:
        # Request failed
        print('Error:', response.status_code)
```

```python
def get_data_fun_params(url,activity,team):

    """ This function will fecth the data from api url and return the data"""

    import requests
    from requests.packages.urllib3.exceptions import InsecureRequestWarning

    # Disable SSL certificate verification warning
    requests.packages.urllib3.disable_warnings(InsecureRequestWarning)

    params ={
        'activity':activity,
        'teamName':team
    }

    # Make a GET request to the API endpoint
    response = requests.get(url,verify=False,params=params)

    # Check the response status code
    if response.status_code == 200:
        # Request was successful
        data = response.json()
        # Process the data as needed
        return data
    else:
        # Request failed
        print('Error:', response.status_code)
```

```python
for team in teams:
    for activity in act_list:
        data = get_data_fun_params("https://vlgspfitcdx.devsys.net.sap:4430/point",activity,team)
        
        for dict_item in data:
            new_dict = dict_item
            new_dict['activity'] = activity
            new_dict['team'] = team
            activity_dict.append(new_dict)
    
        
```
